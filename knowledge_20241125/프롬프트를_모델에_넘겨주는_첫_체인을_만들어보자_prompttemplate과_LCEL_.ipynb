````python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain.schema.runnable import RunnablePassthrough

# LLM model ìƒì„±
model = ChatOpenAI(model="gpt-4o-mini")

# prompt í…œí”Œë¦¿ ìƒì„±
prompt = ChatPromptTemplate.from_template("{dog_breeds_input}ë“¤ì˜ ê³ í–¥ì€ ì–´ë””ì•¼?")

# RunnablePassthrough ê°ì²´ ìƒì„±
passthrough = RunnablePassthrough()

# chain ìƒì„±
chain = prompt | model | passthrough | passthrough | StrOutputParser()

# chain ì‹¤í–‰
answer = chain.invoke({"dog_breeds_input": "ë§í‹°ì¦ˆ"})
print(answer)



#inputê³¼ typeì´ ë§ì•„ì•¼í•œë‹¤.
#RunnablePassthrough í´ë˜ìŠ¤ì´ê³  ê°ì²´ë¼ì„œ, prompt íƒ€ì… ìƒê´€ì—†ì´ ë“¤ì–´ì˜¬ ìˆ˜ ìˆëŠ”ë°, ë§ì§€ ì•ŠëŠ” ê³„ì‚°í•˜ë ¤ê³  í•˜ë©´ ì—ëŸ¬..

# ê²€ì¦ìš©ë„, ì¤‘ê°„ì— ì˜¤ë˜ ê±¸ë¦¬ë©´ ê±°ê¸°ê°€ ë¬¸ì œ -> ê·¸ìª½ë¶€ë¶„ ë””ë²„ê¹…í•´ì„œ í•´ê²°í•˜ê¸°
````

````python
import time
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain.schema.runnable import RunnablePassthrough

# LLM model ìƒì„±
model = ChatOpenAI(model="gpt-4o-mini")

# prompt í…œí”Œë¦¿ ìƒì„±
prompt = ChatPromptTemplate.from_template("{dog_breeds_input}ë“¤ì˜ ê³ í–¥ì€ ì–´ë””ì•¼?")

# ì²˜ë¦¬ ì‹œê°„ì„ ê¸°ë¡í•˜ëŠ” Custom RunnablePassthrough
class TimingRunnablePassthrough(RunnablePassthrough):
    def invoke(self, input, *arg):
        start_time = time.time()  # ì‹œì‘ ì‹œê°„ ê¸°ë¡
        output = super().invoke(input)  # ê¸°ì¡´ Passthrough ë™ì‘ ìˆ˜í–‰
        end_time = time.time()  # ì¢…ë£Œ ì‹œê°„ ê¸°ë¡

        # ì²˜ë¦¬ ì‹œê°„ ì¶œë ¥
        processing_time = end_time - start_time
        print(f"Processing time: {processing_time:.6f} seconds")

        return output

# Custom RunnablePassthrough ê°ì²´ ìƒì„±
timed_passthrough = TimingRunnablePassthrough()

# chain ìƒì„±
chain = prompt | timed_passthrough | model | timed_passthrough | StrOutputParser() | timed_passthrough

# chain ì‹¤í–‰
answer = chain.invoke({"dog_breeds_input": "ë§í‹°ì¦ˆ"})
print(answer)
````
